```python
from kaggle_environments import make

import main
from tf_agents_testcases.dqn import DQNet
```

```python
agent = DQNet("gym_halite:halite-v0")
```

```python
returns, policy = agent.train()
```

```python
halite_agent = main.get_halite_agent(policy)
```

```python
# these are the settings used for a halite v0.1
board_size = 5
starting_halite = 5000

env = make("halite",
           configuration={"size": board_size,
                          "startingHalite": starting_halite},
           debug=True)
```

```python
env.run([halite_agent])
env.render(mode="ipython", width=800, height=600)
```
